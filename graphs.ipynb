```python
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i, y[i], y[i])
```

```python
X = ['GA','GAB-A','GAB-B','GAB-C1', 'GAB-C2']
max_avg_fitness = [888, 905, 917, 912, 784]
max_avg_time = [18, 19, 43, 48, 33]
max_avg_FEvals = [2139, 2146, 2146, 202859, 214353]

X_axis = np.arange(len(X))

plt.bar(X_axis - 0.2, max_avg_fitness, 0.4, label = 'Max Average Fitness', color = 'b')
plt.bar(X_axis + 0.2, max_avg_time, 0.4, label = 'Max Average Time (s)', color = 'g')
#plt.bar(X_axis + 0.4, max_avg_FEvals, 0.4, label = 'Max Average FEvals', color = 'y')

addlabels(X, max_avg_fitness)
addlabels(X, max_avg_time)

plt.xticks(X_axis, X)
plt.xlabel("Experimentation Groups")
plt.ylabel("Performance")
# plt.title("Number of Students in each group")
plt.legend(loc = 'center')
plt.show()
```
